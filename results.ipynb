```python
import numpy as np
import re
import os
import pickle
import matplotlib.pyplot as plt

def extract_generation(filename):
    match = re.search(r'gen(\d+)', filename)
    if match:
        return int(match.group(1))
    else:
        return -1
    
directory = 'stats/norm-phero-test/'   
filenames = os.listdir(directory)
sorted_filenames = sorted(filenames, key=extract_generation)
fitness = []
for gen in sorted_filenames:
    with open(directory + gen, 'rb') as output:
        population_fitness = pickle.load(output)
        fitness.append(np.max(population_fitness))
window_size = 9
padded_fitness = np.pad(fitness, (window_size // 2, window_size // 2), mode='edge')
smoothed_fitness = np.convolve(padded_fitness, np.ones(window_size) / window_size, mode='valid')
x = range(len(sorted_filenames))
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.plot(x, smoothed_fitness, 'y-', label='Using the whole phero matrix')


################################################


directory = 'stats/norm-phero-rand/'   
filenames = os.listdir(directory)
sorted_filenames = sorted(filenames, key=extract_generation)
fitness = []
for gen in sorted_filenames:
    with open(directory + gen, 'rb') as output:
        population_fitness = pickle.load(output)
        fitness.append(np.max(population_fitness))
window_size = 3
padded_fitness = np.pad(fitness, (window_size // 2, window_size // 2), mode='edge')
smoothed_fitness = np.convolve(padded_fitness, np.ones(window_size) / window_size, mode='valid')
x = range(len(sorted_filenames))
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.plot(x, smoothed_fitness, 'b-', label='Random')


################################################


directory = 'stats/mod-test/'   
filenames = os.listdir(directory)
sorted_filenames = sorted(filenames, key=extract_generation)
fitness = []
for gen in sorted_filenames:
    with open(directory + gen, 'rb') as output:
        population_fitness = pickle.load(output)
        fitness.append(np.max(population_fitness))
window_size = 5
padded_fitness = np.pad(fitness, (window_size // 2, window_size // 2), mode='edge')
smoothed_fitness = np.convolve(padded_fitness, np.ones(window_size) / window_size, mode='valid')
x = range(len(sorted_filenames))
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.plot(x, smoothed_fitness, 'r-', label='Only neighbors phero vector')

################################################




directory = 'stats/mod-hidden-large/'   
filenames = os.listdir(directory)
sorted_filenames = sorted(filenames, key=extract_generation)
fitness = []
for gen in sorted_filenames:
    with open(directory + gen, 'rb') as output:
        population_stats = pickle.load(output)
        population_fitness = population_stats['fitness']
        visited_points = population_stats['point_visited']
        max_fit = np.max(population_fitness) - visited_points[np.argmax(population_fitness)]*10 + 50
        fitness.append(max_fit)
window_size = 1
padded_fitness = np.pad(fitness, (window_size // 2, window_size // 2), mode='edge')
smoothed_fitness = np.convolve(padded_fitness, np.ones(window_size) / window_size, mode='valid')
x = range(len(sorted_filenames))
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.plot(x, smoothed_fitness, 'c--', label='Pre initialized hidden layers in FFNN with global phero update')



################################################


directory = 'stats/mod-hidden-large-noglobal/'   
filenames = os.listdir(directory)
sorted_filenames = sorted(filenames, key=extract_generation)
fitness = []
for gen in sorted_filenames:
    with open(directory + gen, 'rb') as output:
        population_stats = pickle.load(output)
        population_fitness = population_stats['fitness']
        visited_points = population_stats['point_visited']
        max_fit = np.max(population_fitness) - visited_points[np.argmax(population_fitness)]*10 + 50
        fitness.append(max_fit)
window_size = 1
padded_fitness = np.pad(fitness, (window_size // 2, window_size // 2), mode='edge')
smoothed_fitness = np.convolve(padded_fitness, np.ones(window_size) / window_size, mode='valid')
x = range(len(sorted_filenames))
plt.xlabel('Generations')
plt.ylabel('Fitness')
plt.plot(x, smoothed_fitness, 'c', label='Pre initialized hidden layers in FFNN')

plt.axhline(y=150-65, color='black', linestyle='--', label='Average randomwalk')
plt.axhline(y=150-45, color='g', linestyle='--', label='Average standard ACO')
#plt.legend()
plt.show()
```
